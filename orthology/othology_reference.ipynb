```python
human_genes = pd.read_csv('human_genes_biomart.csv')
human_genes.dropna(axis=0, inplace=True)
print(f'Total genes: {len(human_genes)}')
print(f"Unique gene names: {len(human_genes['gene_name'].unique())}")
print(f"Repeated genes: {len(human_genes)-len(human_genes['gene_name'].unique())}")

human_map = {}
human_ch_map = {}
for i in range(len(human_genes)):
    if human_genes['gene_name'].iloc[i]!=np.nan:
        human_map[human_genes['gene_id'].iloc[i]]= human_genes['gene_name'].iloc[i]
    if human_genes['chromosome'].iloc[i]!=np.nan:
        human_ch_map[human_genes['gene_id'].iloc[i]]= human_genes['chromosome'].iloc[i]

human_map
```

```python
human_genes
```

```python
nmr_genes = pd.read_csv('nmr_genes_biomart.csv')
nmr_genes.dropna(axis =0, inplace=True)
print(f'Total genes: {len(nmr_genes)}')
print(f"Unique gene names: {len(nmr_genes['gene_name'].unique())}")
print(f"Repeated genes: {len(nmr_genes)-len(nmr_genes['gene_name'].unique())}")

nmr_map = {}
nmr_ch_map = {}
nmr_name2acc_map = dict(zip(human_genes['gene_name'], human_genes['gene_id']))
for i in range(len(nmr_genes)):
    if nmr_genes['gene_name'].iloc[i]!=np.nan:
        nmr_map[nmr_genes['gene_id'].iloc[i]]= nmr_genes['gene_name'].iloc[i]
    if nmr_genes['chromosome'].iloc[i]!=np.nan:
        nmr_ch_map[nmr_genes['gene_id'].iloc[i]]= nmr_genes['chromosome'].iloc[i]
nmr_map
```

```python

human_map
```

```python
nmr2human['human_gene_name'] = nmr2human['homo_sapiens_gene'].map(human_map)
nmr2human.rename(columns={'heterocephalus_glaber_female_gene':'nmr_gene_name','homo_sapiens_gene':'human_gene_acc'}, inplace=True)
```

```python
human2nmr['nmr_gene_name'] = human2nmr['heterocephalus_glaber_female_gene'].map(nmr_map)
human2nmr.rename(columns={'heterocephalus_glaber_female_gene':'nmr_gene_acc','homo_sapiens_gene':'human_gene_name'}, inplace=True)
human2nmr
```

```python
print(len(human2nmr[human2nmr['nmr_gene_name'].notna()]))
print(len(nmr2human[nmr2human['human_gene_name'].notna()]))
```

```python
merged = human2nmr.merge(nmr2human,how='outer', on=['human_gene_name','orthology_type'], suffixes=('_h2n', '_n2h'))
```

```python
merged
```

```python
final = merged[merged['perc_id_h2n'].notna() & merged['perc_id_n2h'].notna()].copy()
final['perc_id']=final[['perc_id_h2n','perc_id_n2h']].max(axis=1)
final.rename(columns={'nmr_gene_name_h2n':'nmr_gene_name'}, inplace=True)
final = final[['human_gene_name','human_gene_acc', 'nmr_gene_name', 'nmr_gene_acc', 'perc_id', 'orthology_type']]
final
```

```python
temp1 = merged[merged['perc_id_h2n'].notna() & merged['perc_id_n2h'].isna()].copy()
temp1.rename(columns={'nmr_gene_name_h2n':'nmr_gene_name', 'perc_id_h2n':'perc_id'}, inplace=True)
temp1 = temp1[['human_gene_name','human_gene_acc', 'nmr_gene_name', 'nmr_gene_acc', 'perc_id', 'orthology_type']]
temp1
```

```python
## Add human gene accession according to its gene name. Consider that this step is not perfect as some genes names correspond to many accessions. 
human_name2acc_map = dict(zip(human_genes['gene_name'], human_genes['gene_id']))
temp1['human_gene_acc'] = temp1['human_gene_name'].map(human_name2acc_map)
temp1
```

```python
temp2 = merged[merged['perc_id_h2n'].isna() & merged['perc_id_n2h'].notna()].copy()
temp2.rename(columns={'nmr_gene_name_n2h':'nmr_gene_name', 'perc_id_n2h':'perc_id'}, inplace=True)
temp2 = temp2[['human_gene_name','human_gene_acc', 'nmr_gene_name', 'nmr_gene_acc', 'perc_id', 'orthology_type']]
```

```python
nmr_name2acc_map = dict(zip(nmr_genes['gene_name'],nmr_genes['gene_id']))
temp2['nmr_gene_acc'] = temp2['nmr_gene_name'].map(nmr_name2acc_map)
temp2
```

```python
final = pd.concat([final, temp1, temp2], axis=0).drop_duplicates(['human_gene_name','human_gene_acc'])
```

```python
human_name_rept = dict(human_genes.groupby('gene_name').count()['description'].sort_values())
print(human_name_rept)
nmr_name_rept = dict(nmr_genes.groupby('gene_name').count()['description'].sort_values())
print(nmr_name_rept)
```

```python
final['human_name_rep'] = final['human_gene_name'].map(human_name_rept)
final['nmr_name_rep'] = final['nmr_gene_name'].map(nmr_name_rept)
final['human_chr'] = final['human_gene_acc'].map(human_ch_map)
final['nmr_chr'] = final['nmr_gene_acc'].map(nmr_ch_map)
```
