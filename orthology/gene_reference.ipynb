```python
import pandas as pd
import numpy as np
import re
```

```python
gene_reference = pd.read_csv('/home/raquelcr/scanpy/combined_adata.csv')
gene_reference
```

```python
gene_reference['Local_ID'] = 'LOCAL' + gene_reference.index.astype(str).str.zfill(11)
```

```python
mask_human = (gene_reference['Human gene stable ID']!= gene_reference['Human gene name']) & (gene_reference['Human gene stable ID'].notnull())
mask_mouse = (gene_reference['Mouse gene stable ID']!= gene_reference['Mouse gene name']) & (gene_reference['Mouse gene stable ID'].notnull())
mask_nmr = (gene_reference['gene_ids']!= gene_reference['NMR gene name']) & (gene_reference['gene_ids'].notnull())
```

```python
# Create global ids with priority nmr > human > mouse 
gene_reference['Global_name'] = np.nan
gene_reference['Global_ID'] = np.nan

for i in range(len(gene_reference)):
    if mask_nmr[i]:
        gene_reference.loc[i, 'Global_name'] = gene_reference.loc[i, 'NMR gene name']
        gene_reference.loc[i, 'Global_ID'] = gene_reference.loc[i, 'gene_ids']
    elif mask_mouse[i]:
        gene_reference.loc[i, 'Global_name'] = gene_reference.loc[i, 'Mouse gene name']
        gene_reference.loc[i, 'Global_ID'] = gene_reference.loc[i, 'Mouse gene stable ID']
    elif mask_human[i]:
        gene_reference.loc[i,'Global_name']= gene_reference.loc[i, 'Human gene name']
        gene_reference.loc[i, 'Global_ID'] = gene_reference.loc[i, 'Human gene stable ID']
    else:
        gene_reference.loc[i, 'Global_name'] = gene_reference.loc[i, 'gene_ids']
        gene_reference.loc[i, 'Global_ID'] = gene_reference.loc[i, 'gene_ids']

gene_reference
```

```python
print(len(set(gene_reference['Global_ID'])))
print(len(gene_reference))
len(set(gene_reference['Global_ID'])) == len(gene_reference)
```

```python
print(len(set(gene_reference['Global_name'])))
print(len(gene_reference))
len(set(gene_reference['Global_name'])) == len(gene_reference)
print(f'Difference: {len(gene_reference) - len(set(gene_reference["Global_name"]))} duplicated gene names')
# There are some duplicated gene names, but not gene ids
```

```python
unique_genes = gene_reference.groupby('Global_name').count()['Global_ID']
not_unique = unique_genes[unique_genes>1].index.tolist()
duplicated_genes = gene_reference[gene_reference['Global_name'].isin(not_unique)].sort_values('Global_name')   
duplicated_genes
```
